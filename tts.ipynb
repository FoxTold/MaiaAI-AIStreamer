```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch
pipeline = KPipeline(lang_code='a')

```

```python
text = '''
WOOHOO! Welcome to the stream, Kevin! Thanks for joining us, and I hope you're ready for some epic Minecraft adventures! Thanks to all our new followers for choosing to hang out with me today! Let's get this stream started! '''
generator = pipeline(text, voice='af_heart')
for i, (gs, ps, audio) in enumerate(generator):
    print(i, gs, ps)
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{i}.wav', audio, 24000)
```
